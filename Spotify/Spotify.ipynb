In my Spotify wrapped from 2024 my top artists were Mt. Joy, Mac Miller and Tyler the Creator



These are all of Tyler, The Creator, Mac Miller and Mt. Joy's songs on spotify

```python
import pandas as pd
spotify = pd.read_csv('https://bcdanl.github.io/data/spotify_all.csv')
spotify.set_index('artist_name', inplace=True)
Tyler = spotify.loc['Tyler, The Creator',['track_name','album_name']]
Tyler
```

```python
Mac = spotify.loc['Mac Miller',['track_name','album_name']]
Mac
```

```python
Mt = spotify.loc['Mt. Joy',['track_name','album_name']]
Mt
```
